```python
# 2. FTP library에서 FTP 가져오기
from ftplib import FTP
```

```python
# FTP(접속 가능한 사이트 주소) 입력
ftp = FTP('ftp1.at.proftpd.org')
```

```python
# ftp 익명 로그인 가능한지 확인
ftp.login()
```

```python
# 해당 ftp 서버의 파일 목록 가져오기
ftp.retrlines('LIST')
```

```python
# ftp 접속 끊기
ftp.quit()
```
